* http://localhost:8100/docs#/default/transcribe_audio_file_transcribe__post

````python
import requests

# 要上传文件的目标URL
url = 'http://localhost:8100/transcribe/'

# MP3文件的实际路径
file_path = './file/001.mp3'

# 打开MP3文件并上传
with open(file_path, 'rb') as f:
    files = {'file': (file_path.split('/')[-1], f, 'audio/mp3')}
    response = requests.post(url, files=files)

# 打印服务器的响应
print(response.text)
````
